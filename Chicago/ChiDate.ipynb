```python
import pandas as pd
import numpy as np
```

```python
chi_crimes = pd.read_table('chi_town.txt', delimiter = ',', names =('Date', 'Charge', "Description"))
```

```python
chi_crimes
```

```python
chi_crimes.Date = chi_crimes.Date.str[2:12]
chi_crimes.Date
```

```python
chi_crimes['Date'] = pd.to_datetime(chi_crimes.Date)
chi_crimes['Date'] = chi_crimes['Date'].dt.strftime('%m/%d/%Y')
chi_crimes.Date = chi_crimes.Date.str.replace('2015', '15')

chi_crimes.head(5)
```

```python
chi_crimes.to_csv("Chi_crime_cleaned.csv")
```
