```python
#Dependancies
import pandas as pd
```

```python
#Locate Portland Crime CSV
csvFile = "Resources/RAW/Portland_crime_data.csv"

#Read Portlnd Crime CSV
crimePortland = pd.read_csv(csvFile)
crimePortland.head()
```

```python
#Display only occur date and type of offense
modifiedData = crimePortland[["Occur Date","Offense Type"]]
modifiedData.head()
```

```python
#Extract year of occur date 
modifiedData['Year'] = pd.DatetimeIndex(modifiedData['Occur Date']).year
modifiedData.head()
```

```python
#Drop all data in years except 2015
cleanedData = modifiedData.loc[modifiedData["Year"] == int("2015"), :]
cleanedData
```

```python
#Display wanted columns
finalData = cleanedData[["Occur Date","Offense Type"]]
finalData
```

```python
#Save as CSV
finalData.to_csv("Resources/Cleaned_Portland_Crime.csv")
```
