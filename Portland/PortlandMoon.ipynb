```python
# Import Dependencies
import pandas as pd
import numpy as np
```

```python
# define file path
csv_path = "Cleaned_Portland_Crime.csv"
portland_df = pd.read_csv(csv_path)


portland_df.head()
```

```python
day = portland_df['Occur Date'].value_counts()

day
```

```python
crime_date = pd.DataFrame(day)
crime_date.columns=['Crime Count']
crime_date.index.name = "Date"
crime_cleaned = crime_date.sort_values("Date", ascending = True)

crime_cleaned.head()
```

```python
# define file path
csv_path = "MoonPhase.csv"
moon_df = pd.read_csv(csv_path)
moon_df.head(5)
```

```python
# combine both of the csv files
portland_moon_df = pd.merge(crime_cleaned, moon_df, how="outer", on = "Date")

portland_moon_df
```

```python
pdx_crime_group = portland_moon_df.set_index('Date')

pdx_crime_group.head(10)
```

```python
# Export as csv
pdx_crime_group.to_csv("PdxMoon.csv")
```
