```python
import pandas as pd
import numpy as np
```

```python
#Bring in Atlanta Data
atl_path = "Atlanta/AtlantaCrimeSummary"
atlanta_df = pd.read_csv(atl_path)
data = atlanta_df.rename(columns={"Type of Crime":"Crime"})
print(len(atlanta_df))
data.head()
```

```python
#Get a list of Crime Types
crime_ls = pd.unique(data["Crime"])
print(len(crime_ls))
print(crime_ls) 
```

```python
#categorize the crimes as violent and non violent
violent = []
violent_count = 0
nonvio_count = 0
for index, row in data.iterrows():
    #print(index)
    #print(row[1])
    
    if (row[1] == crime_ls[0]) | (row[1] == crime_ls[1]) | (row[1] == crime_ls[3]) | (row[1] == crime_ls[8]) | (row[1] == crime_ls[9]):
        violent.append("violent")
        violent_count += 1
    else:
        violent.append("non-violent")
        nonvio_count += 1
print("Violent: " + str(violent_count)) 
print("Non-Violent: " + str(nonvio_count))
data["Category"] = violent
data.head()
```

```python
#group data and make new dataframes for violen and non violent crimes in a day 
grouped = data.groupby(['Date of Crime', "Category"])
atl_grouped = grouped.count()

vio_count = []
nonv_count = []
v_date = []
nonv_date = []

#loop through and count violent versus nonviolent indexes
for index, inner in atl_grouped.iterrows():
    #print("index: " + str(index[1]))
    #print("inner: " + str(inner[0]))
    if index[1] == "violent":
        vio_count.append(inner[0])
        v_date.append(index[0])
    else:
        nonv_count.append(inner[0])
        nonv_date.append(index[0])
        
    

zip_vd = list(zip(v_date, vio_count))
zip_nvd = list(zip(nonv_date, nonv_count))
violent_df = pd.DataFrame(zip_vd, columns = ["Date of Crime", "Violent Crime"]) 
nonv_df = pd.DataFrame(zip_nvd, columns = ["Date of Crime", "NonViolent Crime"])
merge_v = pd.merge(violent_df, nonv_df, on="Date of Crime", how="outer")
merge_v.head()
```

```python
#merge violent nonviolent dataframe with crime count data frame
group_ondate =data.groupby(['Date of Crime'])
group_count = group_ondate.count()
merge_atl = pd.merge(group_count, merge_v, on = 'Date of Crime', how = 'outer')
merge_atl = merge_atl.drop(['Category'], axis=1)
merge_atl.head()
```

```python
grouped_data = port_df.groupby(['Occur Date', "affliction"])
final_port = grouped_data.count()

violent_count = []
nviolent_count = []
vio_date = []
nv_date = []
#loop through and count violent versus nonviolent indexes
for index, inner in final_port.iterrows():

    if index[1] == "violent":
        violent_count.append(inner[1])
        vio_date.append(index[0])
    else:
        nviolent_count.append(inner[1])
        nv_date.append(index[0])
        

    

list_vd = list(zip(vio_date, violent_count))
list_nvd = list(zip(nv_date, nviolent_count))
vio_df = pd.DataFrame(list_vd, columns = ["Occur Date", "Violent Crime"]) 
non_df = pd.DataFrame(list_nvd, columns = ["Occur Date", "NonViolent Crime"])
merge_vnv = pd.merge(vio_df, non_df, on="Occur Date", how="outer")
merge_vnv.head()
```

```python
#merge violent nonviolent dataframe with crime count data frame
group_pdate =port_df.groupby(['Occur Date'])
group_pcount = group_pdate.count()
merge_pdx = pd.merge(group_pcount, merge_vnv, on = 'Occur Date', how = 'outer')
merge_pdx = merge_pdx.drop(['Unnamed: 0', 'affliction'], axis=1)
merge_pdx = merge_pdx.fillna(0)
merge_pdx.head()
```
