```python
# Import Dependencies
import pandas as pd
import numpy as np
```

```python
# define file path
csv_path = "Resources/AtlantaCrime2015.csv"
atlanta_df = pd.read_csv(csv_path)

atlanta_df.head()
```

```python
crimes = atlanta_df['UCR Literal'].unique()
crimes
```

```python
date = atlanta_df['Occur Date']
crime = atlanta_df['UCR Literal']

summary_atl = pd.DataFrame({"Date" : date,
                           "Type of Crime " : crime})

summary_atl.head()
```

```python
date_df = pd.DataFrame({"Date": date})
date_df.head()
```

```python
# define file path
csv_path = "Resources/MoonPhase.csv"
moon_df = pd.read_csv(csv_path)
moon_df.head(5)

```

```python
moon_df.Date = moon_df.Date.str.replace('2015', '15')
moon_df.head(5)
```

```python
day = summary_atl['Date'].value_counts()

day
```

```python
crime_date = pd.DataFrame(day)
crime_date.columns=['Crime Count']
crime_date.index.name = "Date"
crime_cleaned = crime_date.sort_values("Date", ascending = True)

crime_cleaned.head()
```

```python
# combine both of the csv files
Atlanta_moon_df = pd.merge(crime_cleaned, moon_df, how="outer", on = "Date")

Atlanta_moon_df
```

```python
crime_group = Atlanta_moon_df.set_index('Date').groupby(['Moon Phase'])

crime_group.head()
```
