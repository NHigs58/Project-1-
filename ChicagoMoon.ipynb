```python
# Import Dependencies
import pandas as pd
import numpy as np
```

```python
# Chicago Data
csv_path = "chi_crime_cleaned.csv"
chi_df = pd.read_csv(csv_path)
chi_df.head()

```

```python
date = chi_df['Date']
crime = chi_df['Charge']

summary_chi = pd.DataFrame({"Date" : date,
                           "Type of Crime " : crime})



summary_chi.Date = summary_chi.Date.str.replace('01', '1')
summary_chi.Date = summary_chi.Date.str.replace('02', '2')
summary_chi.Date = summary_chi.Date.str.replace('03', '3')
summary_chi.Date = summary_chi.Date.str.replace('04', '4')
summary_chi.Date = summary_chi.Date.str.replace('05', '5')
summary_chi.Date = summary_chi.Date.str.replace('06', '6')
summary_chi.Date = summary_chi.Date.str.replace('07', '7')
summary_chi.Date = summary_chi.Date.str.replace('08', '8')
summary_chi.Date = summary_chi.Date.str.replace('09', '9')

summary_chi.head()
```

```python
day = summary_chi['Date'].value_counts()

day
```

```python
crime_date = pd.DataFrame(day)
crime_date.columns=['Crime Count']
crime_date.index.name = "Date"
crime_cleaned = crime_date.sort_values("Date", ascending = True)

crime_cleaned.head()
```

```python
# define file path
csv_path = "Resources/MoonPhase.csv"
moon_df = pd.read_csv(csv_path)
moon_df.head()
```

```python
moon_df.Date = moon_df.Date.str.replace('2015', '15')
moon_df.head()
```

```python
# combine both of the csv files
chi_moon_df = pd.merge(crime_cleaned, moon_df, how="outer", on = "Date")

chi_moon_df
```

```python
chi_crime_group = chi_moon_df.set_index('Date')
chi_crime_group
```

```python
# Export as csv
chi_crime_group.to_csv("ChiMoon.csv")
```
