```python
# Import Dependencies
import pandas as pd
import numpy as np
```

```python
# define file path
csv_path = "Resources/AtlantaCrime2015.csv"
atlanta_df = pd.read_csv(csv_path)

atlanta_df.head()
```

```python
date = atlanta_df['Occur Date']
crime = atlanta_df['UCR Literal']

summary_atl = pd.DataFrame({"Date of Crime" : date,
                           "Type of Crime " : crime})

summary_atl.head()
```

```python
summary_atl.to_csv("AtlantaCrimeSummary", index=False, header=True)
```
